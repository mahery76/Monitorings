Simple graph from tutorial

```python
# from matplotlib.pyplot import plot, show, title, xlabel, ylabel, axhline, legend
import matplotlib.pyplot as plt
sunny_days = [8, 10, 12, 15, 9, 11, 14, 13, 16, 12, 10, 9]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
avg_sunny_days = sum(sunny_days) / len(sunny_days)
plt.plot(months, sunny_days, marker='o', linestyle='-', color='orange', label='Sunny Days')
plt.title("Sunny days by month", fontsize=20)
plt.xlabel("Month", fontsize=14)
plt.ylabel("Sunny days", fontsize=14)
plt.axhline(avg_sunny_days, color='blue', linestyle='--', label='Average Sunny Days')
# this will show the label from plot and axhline 
plt.legend() 

plt.show()
```

ping 8.8.8.8 -t
- Store the time response from ping into an array pingResults
- display the pingResults


```python
import subprocess
import re
import matplotlib.pyplot as plt

pingResults = []

process = subprocess.Popen( 
    ["ping", "8.8.8.8", "-t"],
    stdout=subprocess.PIPE,
    text=True
)

for line in process.stdout:
    # Extract time value using regex and works for = or < symbols
    match = re.search(r'(time|temps)[=<]\s*([\d.]+)', line)
    if match:
        time_value = float(match.group(2))
        pingResults.append(time_value)
        print(pingResults, flush=True)

```
